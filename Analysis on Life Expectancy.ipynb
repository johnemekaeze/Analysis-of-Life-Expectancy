```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the csv file and displaying the first 20 rows
df = pd.read_csv('Life_Expectancy_Data.csv')
df.head(20)
```

```python
# Displaying information about the data
df.info()
```

```python
df.columns
```

```python
# Correcting the column names
df = df.rename(columns = {'Life expectancy ':'Life Expectancy',
'infant deaths':'Infant Deaths',
'percentage expenditure':'Percentage Expenditure',
'Measles ':'Measles',
' BMI ':'BMI',
'under-five deaths ':'Under-Five Deaths',
'Total expenditure':'Total Expenditure',
'Diphtheria ':'Diphtheria',
' HIV/AIDS':'HIV/AIDS',
' thinness  1-19 years':'Thinness (1-19 years)',
' thinness 5-9 years':'Thinness (5-9 years)',
'Income composition of resources':'Income Composition'})
df.head()
```

```python
# Displaying the number of null values for each column 
df.isnull().sum()
```

```python
# Importing the SimpleImputer class
from sklearn.impute import SimpleImputer
```

```python
# Changing the data type of 'Year' column to string
df['Year'] = df['Year'].astype(str)
```

```python
# Separating numeric and non-numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
non_numeric_columns= df.select_dtypes(exclude=['float64', 'int64']).columns.tolist()
```

```python
numeric_columns
```

```python
# Imputing missing values within each country for numeric columns
for country in df['Country'].unique():
    for column in numeric_columns:
        country_data = df[df['Country'] == country][column]
        mean_value = country_data.mean()
        df.loc[df['Country'] == country, column] = df.loc[df['Country'] == country, column].fillna(mean_value)
```

```python
# Checking if there are any missing values left
df.isnull().sum()
```

```python
# Applying overall mean imputation to remaining missing values

# Initiating SimpleImputer to fill the null values with the mean
imputer = SimpleImputer(missing_values = np.nan, strategy ='mean')

# Fitting and transforming numeric data
df_numeric = pd.DataFrame(imputer.fit_transform(df[numeric_columns]), columns=numeric_columns)
```

```python
df_numeric.isnull().sum()
```

```python
# Combining imputed numeric data with non-numeric data

df_imputed = pd.concat([df[non_numeric_columns], df_numeric], axis=1)

df_imputed.head()
```

```python
# Checking for missing values
df_imputed.isnull().sum()
```

```python
# Overwriting the original dataframe with the imputed dataframe
df = df_imputed
```

```python
# Getting more details of the numeric columns including the count, mean, standard deviation, quartiles and extrema for the numeric the count, mean, standard deviation, quartiles and extrema
df.describe()
```

### Visualization

```python
# Calculating the correlation matrix
correlation_matrix = df[numeric_columns].corr()

# Plotting the heatmap
plt.figure(figsize=(10, 8), facecolor='white')
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numeric Columns')
```

```python
# Plotting the histogram of Life Expectancy
plt.figure(figsize=(10, 7), facecolor='white')
plt.hist(df['Life Expectancy'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Life Expectancy')
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
```

```python
# Creating a violin plot comparing life expectancy in developed and developing countries
plt.figure(figsize=(10, 7), facecolor='white')
plt.subplot(1, 2, 1)
sns.violinplot(x='Status', y='Life Expectancy', data=df[df['Status']=='Developed'], palette='Set2')
plt.title('Life Expectancy in Developed Countries')
plt.xlabel('Status')
plt.ylabel('Life Expectancy (years)')

plt.subplot(1, 2, 2)
sns.violinplot(x='Status', y='Life Expectancy', data=df[df['Status']=='Developing'], palette='Set1')
plt.title('Life Expectancy in Developing Countries')
plt.xlabel('Status')
plt.ylabel('Life Expectancy (years)')
```

```python
# Checking all the different years in the data
df.Year.unique()
```

```python
# Filter the dataset for the countries of interest and the years 2000 to 2015
filtered_data = df[(df['Country'].isin(['Canada', 'United Kingdom of Great Britain and Northern Ireland', 'United States of America']))]

# Creating the line plot
plt.figure(figsize=(10, 6), facecolor='white')
for country in ['Canada', 'United Kingdom of Great Britain and Northern Ireland', 'United States of America']:
    country_data = filtered_data.sort_values(by='Year')[filtered_data.sort_values(by='Year')['Country'] == country]
    plt.plot(country_data['Year'], country_data['Life Expectancy'], label=country)

plt.title('Life Expectancy from 2000 to 2015')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.legend()
plt.grid(True)
```

```python
# Verifying that United Kingdom data and that of its constituent countries is not in the dataframe
df.Country.unique()
```

```python
# Filter for the specified countries
countries = ['Belgium', 'Brazil', 'Cameroon', 'Canada', 'China', 'France', 'Ghana', 'India', 'United Kingdom of Great Britain and Northern Ireland', 'United States of America']
filtered_data = df[df['Country'].isin(countries)]

# Proceeding with the analysis
avg_data = filtered_data.groupby('Country').agg({'Infant Deaths': 'mean', 'Life Expectancy': 'mean'}).reset_index()

# Plotting
plt.figure(figsize=(10, 6), facecolor='white')
plt.scatter(avg_data['Infant Deaths'], avg_data['Life Expectancy'], color='blue', alpha=0.5)
for i, txt in enumerate(avg_data['Country']):
    plt.annotate(txt, (avg_data['Infant Deaths'][i], avg_data['Life Expectancy'][i]))
plt.title('Average Infant Deaths vs. Life Expectancy')
plt.xlabel('Average Infant Deaths')
plt.ylabel('Average Life Expectancy')
plt.grid(True)
```

```python
# Filter the data for the year 2012
df_2012= df[df['Year'] == '2012']

# Scatter plot of 'Life expectancy' vs 'Schooling'
plt.figure(figsize=(10, 6), facecolor='white')
plt.scatter(df_2012['Schooling'], df_2012['Life Expectancy'], color='blue')
plt.title('Life Expectancy vs Schooling in 2012')
plt.xlabel('Schooling (years)')
plt.ylabel('Life Expectancy (years)')
plt.grid(True)
```

```python
# Calculating the correlation
df['Schooling'].corr(df['Life Expectancy'])
```

```python
# importing the pearsonr class from scipy.stats
from scipy.stats import pearsonr

# Calculate the Pearson correlation
pearson_corr = pearsonr(df_2012['Schooling'], df_2012['Life Expectancy'])
print('Pearson correlation coefficient:', pearson_corr[0])
```

```python
# Importing the LinearRegression class
from sklearn.linear_model import LinearRegression

# Fit a linear regression model
X = df_2012[['Schooling']].values
y = df_2012['Life Expectancy'].values
model = LinearRegression()
model.fit(X, y)

# Predict y values
y_pred = model.predict(X)

# Plot the regression line on the same plot
plt.figure(figsize=(10, 6), facecolor='white')
plt.scatter(df_2012['Schooling'], df_2012['Life Expectancy'], color='blue', alpha=0.5)
plt.plot(df_2012['Schooling'], y_pred, color='red')
plt.title('Life Expectancy vs Schooling in 2012 with Regression Line')
plt.xlabel('Schooling (years)')
plt.ylabel('Life Expectancy (years)')
plt.grid(True)
```
